# Spliting The Dataset

```python
X = df.iloc[:,[2,5,6,7,67]]

X=pd.DataFrame(X)

#Dependant Variable

y =df.iloc[:,4].values

y=pd.DataFrame(y)
```

# Checking For Null Values And Filling The Null Values

```python
X.isnull().sum()
```

```python
X.fillna(X.mean(),inplace=True)
```

```python
X.isnull().sum()
```

```python
y.fillna(y.mean(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
X['Country']=label_encoder.fit_transform(X['Country'])
```

# Dividing The Model Into Train And Test Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=74)
```

```python
x_train
```

```python
y_train
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
```

```python
print("Train: ",reg.score (x_train,y_train)) 
print("Test: ",reg.score (x_test,y_test))
```

```python
y_pred=reg.predict(x_test)
y_pred
```

# Saving the Model

```python
import pickle
pickle.dump(reg,open('HDI.pkl','wb'))
```

```python
pickle.dump(label_encoder,open('label_encoder.pkl','wb'))
```
